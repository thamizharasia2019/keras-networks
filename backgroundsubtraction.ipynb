```python
filepath1='./dataset_binary_noback/smear2005A/normal/'

path2='./dataset_binary_noback/smear2005B/normal/'
import os
import cv2

thresh = 100
# print(os.listdir(path))
for filename in os.listdir(filepath1):  

        img=cv2.imread(filepath1+filename)
        w,h=img.shape[:2]

        imag=img
        for i in range(w):
            for j in range(h):
                if img[i,j,2]==255:
                    imag[i,j,0]=255
                    imag[i,j,1]=255
                else:
                    imag[i,j,0]=0
                    imag[i,j,1]=0
                    imag[i,j,2]=0
# 
#         img_grey1= cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
#         ret,thresh_img1 = cv2.threshold(img_grey1, thresh, 255, cv2.THRESH_BINARY)
        cv2.imwrite(path2+filename,imag)
        print(path2+filename +' is written')
    
```
