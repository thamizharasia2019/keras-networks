```python
from imutils import paths
import numpy as np
import imutils

import cv2
import os
```

```python
basepath='./smear2005Mask/'

imagePaths = list(paths.list_images("./smear2005"))
 
for (i, imagePath) in enumerate(imagePaths):
	# extract the person name from the image path
	print("[INFO] processing image {}/{}".format(i + 1,
		len(imagePaths)))
	name = imagePath.split(os.path.sep)[-1].replace('.BMP','-d.bmp')
	dir1 = imagePath.split(os.path.sep)[-2]
	# load the image, resize it to have a width of 600 pixels (while
	# maintaining the aspect ratio), and then grab the image
	# dimensions
	image = cv2.imread(imagePath)
	maskpath='./smear2005/'+ dir1 + '/'+name
	print(maskpath)
	maskimage = cv2.imread(maskpath)
	grayImage = cv2.cvtColor(maskimage, cv2.COLOR_BGR2GRAY)

	(thresh, blackAndWhiteImage) = cv2.threshold(grayImage, 64, 255, cv2.THRESH_BINARY)
	h =int(blackAndWhiteImage.shape[0])

	w= int(blackAndWhiteImage.shape[1])

#print(blackAndWhiteImage[:20])
	for ii in range(h):
		for j in range(w):
			if blackAndWhiteImage[i][j]==255:
				image[ii][j][0]=0
				image[ii][j][1]=0
				image[ii][j][2]=0
	newname = imagePath.split(os.path.sep)[-1].replace('.BMP','-m.bmp')
	newdir=basepath+dir1
	if not os.path.exists(newdir):
		os.makedirs(newdir)
	filename=basepath+ dir1+ '/'+newname
	print(filename)
	cv2.imwrite(filename,image)    
```
